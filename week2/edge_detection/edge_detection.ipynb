## Edge Detection With OpenCV

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
car_img = cv2.imread('./car.png')
cyclist = cv2.imread('./cyclist.jpg')
```

```python
car_gray = cv2.cvtColor(car_img, cv2.COLOR_BGR2GRAY)
cyclist_gray = cv2.cvtColor(cyclist, cv2.COLOR_BGR2GRAY)
```

```python
car_edges = cv2.Canny(car_gray, 100, 200)
cyclist_edges = cv2.Canny(cyclist_gray, 200,250)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(Image.open('./car.png'))
plt.title('Original Image', size=20);plt.xticks([]);plt.yticks([])
plt.subplot(222)
plt.imshow(car_edges, cmap='gray')
plt.title('Edge Image', size=20);plt.xticks([]);plt.yticks([])
plt.subplot(223)
plt.imshow(Image.open('./cyclist.jpg'))
plt.title('Original Image', size=20);plt.xticks([]);plt.yticks([])
plt.subplot(224)
plt.imshow(cyclist_edges, cmap='gray')
plt.title('Edge Image', size=20);plt.xticks([]);plt.yticks([])
```
